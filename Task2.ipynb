```python

from matplotlib import pyplot as plt

res = df.groupby('country')['type'].apply(pd.value_counts).reset_index()

res.columns=['country','type','count']
res = pd.DataFrame(res,columns=['country','type','count'])
res = res.apply(pd.value_counts).reset_index().sort_values(['count'], ascending=False)


res.dropna()
res = res.head(5)

country = res['index'].tolist()
data = res['count'].tolist()
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = country)

# show plot
plt.show()
```
